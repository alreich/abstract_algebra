## Pauli Matrices

```python
sI = np.array([[1, 0],
               [0, 1]],
              dtype=complex)

sX = np.array([[0, 1],
               [1, 0]],
              dtype=complex)

sY = np.array([[0 + 0j, 0 - 1j],
               [0 + 1j, 0 + 0j]],
              dtype=complex)

sZ = np.array([[1,  0],
               [0, -1]],
              dtype=complex)

pauli = [sI, sX, sY, sZ]
```

```python
print(f"sI = \n{sI}\n")
print(f"sX = \n{sX}\n")
print(f"sY = \n{sY}\n")
print(f"sZ = \n{sZ}\n")
```

```python
pauli_dict = {'I': sI, 'X': sX, 'Y': sY, 'Z': sZ}
pauli_dict
```

```python
new_dict = dict()
pauli_values = pauli_dict.values()
for x in pauli_dict:
    for y in pauli_dict:
        prod = pauli_dict[x] @ pauli_dict[y]
        if not any([np.array_equal(prod, arr) for arr in pauli_values]):
            new_dict[x + y] = prod
new_dict
```

```python
pauli_dict = dict(list(pauli_dict.items()) + list(new_dict.items()))
pauli_dict
```
