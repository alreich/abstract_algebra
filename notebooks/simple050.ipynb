## Examples

```python
rps_elements = ['r', 'p', 's']

rps_table = [['r', 'p', 'r'],
             ['p', 'p', 's'],
             ['r', 's', 's']]
```

```python
rps_ct = make_cayley_table(rps_table, rps_elements)
```

```python
z3_elements = ['e', 'a', 'a^2']

z3_table = [[ 'e' ,  'a' , 'a^2'],
            [ 'a' , 'a^2',  'e' ],
            ['a^2',  'e' ,  'a' ]]
```

```python
z3_ct = make_cayley_table(z3_table, z3_elements)
```

```python
about_tables([rps_ct, z3_ct])
```
