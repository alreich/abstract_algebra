## Generators

The method, below, only looks for single elements that generate the whole algebra.

```python
a5.generators()
```

Here's the code for that method.  Generalize it to look for the smallest sets of elements that generator the whole algebra.

```python
def generators(alg):
    """Return a list of individual elements that generate the entire algebra."""
    elemset = set(alg.elements)
    return [x for x in elemset if set(alg.closure([x], False)) == elemset]
```

```python
generators(a5)
```

```python
def closed_subsets_of_elements(alg, divisors_only, include_inverses):
    closed = set()  # Build the result as a set of sets to avoid duplicates
    all_elements = alg.elements
    n = len(all_elements)
    if divisors_only:
        range_ = divisors(n, non_trivial=True)
    else:
        range_ = range(2, n - 1)
    for i in range_:
        # Look at all combinations of elements: pairs, triples, quadruples, etc.
        for combo in it.combinations(all_elements, i):
            # Freezing is required to add a set to a set
            clo = frozenset(alg.closure(list(combo), include_inverses))
            if len(clo) < n:  # Don't include closures consisting of all elements
                closed.add(clo)
    return list(map(lambda x: list(x), closed))
```
