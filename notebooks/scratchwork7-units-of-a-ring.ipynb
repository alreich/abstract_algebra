```python
def units(ring):
    """Return a list of the Ring's units."""
    xs, ys = np.where(ring.mult_table.table == ring.elements.index(ring.one))
    xy_pairs = list(zip(xs, ys))
    return sorted(list({xy[0] for xy in xy_pairs if (xy[1], xy[0]) in xy_pairs}))
```
